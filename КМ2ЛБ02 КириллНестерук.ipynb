## 2.1 Понятие объекта. Атрибуты. Методы. Встроенные типы объектов

***Объект*** *в Python* — это базовая сущность для представления данных.

---

### Классификация объектов

К объектам в Python **относятся**:

- *данные* встроенных типов;
- *встроенные* и *пользовательские* функции;
- *встроенные* и *пользовательские* классы;
- *экземпляры* классов;
- *модули*.

---

### Внутреннее устройство объектов

***Внутренние объекты*** представляют собой **области выделенной памяти** с достаточным пространством для хранения информации об объекте.  

Информация об объекте **делится** на:

| Компонент | Описание |
|-----------|----------|
| *Данные объекта* | значение объекта, состояние объекта |
| *Метаданные* | служебная информация об объекте |

---

### Атрибуты и методы

***Метаданные*** объекта называются ***атрибутами***.  

> *Атрибут* — это имя переменной или имя функции, которое связано с объектом и используется для получения информации об объекте и совершения действий над этим объектом.

Все атрибуты объекта **составляют** *пространство имен* объекта.  

Атрибуты, которые начинаются и заканчиваются **двумя знаками подчеркивания** (например, `__init__`, `__str__`), чаще всего являются ***методами***, реализующими операции над объектами.

---

### Операции

***Операция*** *(operator)* — это конструкция языка, аналогичная по записи математическим операциям.  

Для обозначения операции часто используют не алфавитно-цифровые, а **специальные символы** (`+`, `-`, `*`, `/`, `==`, `>` и т.д.).

---

## Встроенные типы объектов

Встроенные типы объектов **делятся** на *числа* и *коллекции*.

### Числовые типы

Типы объектов, представляющих **числа**:

| Тип | Описание | Пример |
|-----|----------|--------|
| `int` | целые числа | `42`, `-7`, `1000` |
| `float` | вещественные числа | `3.14`, `-0.001`, `2.0` |
| `complex` | комплексные числа | `3+4j`, `2-1j` |

> Булевый тип (`bool`) **наследуется** от типа `int`, поэтому булевый тип относится к типам числовых объектов.  
> `True` ≈ `1`, `False` ≈ `0`

---

### Коллекции

***Коллекции*** — это **наборы объектов**, которые делятся на три вида:

#### Последовательности (*sequences*)
Позиционно **упорядоченные** коллекции.

| Тип | Описание | Пример |
|-----|----------|--------|
| `str` | строки | `"Hello"`, `'Python'` |
| `list` | списки | `[1, 2, 3]`, `['a', 'b']` |
| `tuple` | кортежи | `(1, 2, 3)`, `('a', 'b')` |

#### Отображения (*mappings*)
**Неупорядоченные** коллекции объектов, хранящихся **по ключам**.

| Тип | Описание | Пример |
|-----|----------|--------|
| `dict` | словарь | `{'name': 'Alice', 'age': 25}` |

#### Множества (*sets*)
**Неупорядоченные** коллекции **не повторяющихся** объектов.

| Тип | Описание | Пример |
|-----|----------|--------|
| `set` | множество | `{1, 2, 3}`, `{'a', 'b'}` |

---

## Литералы и выражения

### Литерал

***Литерал*** *в Python* — это **форма записи данных** для их представления объектом определенного типа.

### Выражение

***Выражение*** *(expression)* в Python — это конструкция языка в виде **комбинации литералов, переменных, операций и вызовов функций**, которая **вычисляется и возвращает значение** в виде объекта.

> Выражение **не хранится** в памяти — в памяти сохраняется только **результирующий объект**.



## Встроенные типы объектов, представляющих числа

***Операция*** *(operator)* — это конструкция языка, аналогичная по записи математическим операциям. Для обозначения операции часто используют не алфавитно-цифровые, а **специальные символы** (`+`, `-`, `*`, `/`, `==`, `>` и т.д.).

### 1. Метод .join()

- **Синтаксис:** `строка_разделитель.join(итерируемый_объект)`
- **Назначение:** Объединяет элементы итерируемого объекта в одну строку
- **Параметры:** `итерируемый_объект` — коллекция, содержащая строки (обязательный)
- **Возвращаемое значение:** Новая строка из элементов коллекции, разделенных разделителем
- **Особенности:** Это метод строки, а не списка

### 2. Метод .split()

- **Синтаксис:** `исходная_строка.split(разделитель, максимальное_количество_разбиений)`
- **Назначение:** Разбивает строку на подстроки по указанному разделителю
- **Параметры:** 
  - `разделитель` — строка-граница (необязательный, по умолчанию пробельные символы)
  - `максимальное_количество_разбиений` — сколько разбиений выполнить (необязательный)
- **Возвращаемое значение:** Список строк, полученных после разбиения
- **Особенности:** Это метод строки, всегда возвращает список

### _2.9 Переменная_




***Переменная*** — это набор символов (имя переменной) для хранения адреса памяти, где находится объект.

**Создание:** переменная создается при первом присваивании значения через оператор `=`: `<переменная> = <значение переменной>`. Значением может быть объект любого типа.

**Типизация:** тип не связан с переменной, а связан со значением. При повторных присваиваниях одной переменной объектов разных типов тип меняется — это называется **динамической типизацией**.

**Ссылочная модель:** переменная хранит не сам объект, а ссылку на него (адрес в памяти). В системной таблице хранится соответствие между именем переменной и адресом объекта. В каждый момент переменная ссылается на один определенный объект.

**Циклические объекты:** если объект коллекции содержит ссылку на самого себя, он называется **циклическим объектом**. При выводе таких объектов Python избегает зацикливания и отображает их специальным образом.

````python
# 1. Циклические объекты
# Список
lst = [1, 2]
lst.append(lst)
print(lst)

# Словарь
d = {'key': 'value'}
d['self'] = d
print(d)

# 2. Количество ссылок
import sys
x = [1, 2, 3]
a = b = c = d = x
print(sys.getrefcount(x))

# 3. Четыре способа копирования списка
original = [1, [2, 3], 4]

copy1 = original.copy()           # метод copy
copy2 = original[:]                # срез
copy3 = list(original)             # конструктор
import copy
copy4 = copy.copy(original)        # модуль copy

# 4. copy vs deepcopy
original = [1, [2, 3], 4]
shallow = copy.copy(original)
deep = copy.deepcopy(original)

original[1][0] = 99

print(f"original: {original}")  # [1, [99, 3], 4]
print(f"shallow:  {shallow}")   # [1, [99, 3], 4]  - вложенный список изменился
print(f"deep:     {deep}")       # [1, [2, 3], 4]   - вложенный список не изменился
````

### _2.10 Включения_

### Списковое включение (list comprehension)

***Списковое включение*** — это выражение, которое создает **НОВЫЙ СПИСОК** за счет выполнения выражения на каждом элементе последовательности (итерируемого объекта), по одному за раз. Выражение спискового включения записывается в квадратных скобках и содержит конструкцию цикла `for ... in ...`.

**Синтаксис:** `[выражение for элемент in итерируемый_объект]`

**Назначение:** применяется, когда необходимо выполнить выражение на каждом элементе итерируемого объекта и получить список в результате.

**Условия:** синтаксис позволяет дополнительно задавать ограничения с использованием конструкции `if`:
`[выражение for элемент in итерируемый_объект if условие]`



````python
# 1. Множество с условием
a = {i for i in range(1, 20) if i % 3 == 0}
print(a)  
````

````python
# 2. Матрица 7×8 через вложенные включения
matrix = [[1 for i in range(8)] for j in range(7)]
for row in matrix:
    print(row)
````

````python
# 3. Матрица 7×8 через numpy
import numpy as np
matrix_np = np.ones((7, 8), dtype=int)
print(matrix_np)
````

````python
# 4. Словарь с двумя уровнями вложенности
d = {i: {j: j**2 for j in range(1, 4)} for i in range(1, 4)}
print(d)
````

### _2.11 Форматированный вывод строк_



Существует три способа форматированного вывода строк на основе подстановок в исходную формальную строку объектов произвольных типов согласно шаблону форматной строки:

- **Бинарная операция `%`:** `'...%s...%d...' % (obj1, obj2)`
- **Строковый метод `format`:** `'...{}.{}.{}'.format(obj1, obj2, obj3)`
- **Форматированный строковый литерал `f-string`:** `f'...{x}...{y}...'`

````python
# Пример с переменной
x = 42

# f-string
print(f"{x=}")           
print(f"{x=:06d}")       

# бинарная операция %
print("x=%s" % x)        
print("x=%06d" % x)      

# метод format
print("{0}={1}".format("x", x))        
print("{0}={1:06d}".format("x", x))    
````

````python
# Пример с выражением
a, b = 7, 15

# f-string
print(f"{a + b=}")          
print(f"{a + b=:06d}")      

# бинарная операция %
print("a + b=%s" % (a + b))       
print("a + b=%06d" % (a + b))      

# метод format
print("{0}={1}".format("a + b", a + b))       
print("{0}={1:06d}".format("a + b", a + b))   
````

### _2.12 Файловый объект_


### 2.12 Файловый объект

***Файловый объект*** — это объект, предоставляющий интерфейс для работы с файлами на диске или устройствами хранения и передачи данных (стандартный поток ввода/вывода, буфер памяти, сокет и др.).

Файловый объект в Python создается вызовом встроенной функции `open(file, mode)`, аргументами которой являются имя файла `file` в виде строки и режим обработки файла `mode` в виде строки:
```python
file = open('tmp.txt', 'rt')
print(file)  # <_io.TextIOWrapper name='tmp.txt' mode='rt' encoding='cp1251'>



Python 3.X проводит четкое различие между текстовыми и двоичными данными в файлах:

- **Текстовые файлы** представляют содержимое в виде строк типа `str`, автоматически выполняя кодирование и декодирование Unicode при записи и чтении данных.
- **Двоичные файлы** представляют содержимое в виде специальных строк `bytes` и позволяют получать доступ к содержимому файла, не изменяя его.

Встроенный тип данных **байтовая строка (bytes)** — это последовательность символов для хранения и представления байтовой информации, используемой в файлах медиа-данных и при передаче через сеть.

````python
# open - открытие файла
f = open('test.txt', 'w')

# write - запись строки
f.write('строка 1\n')
f.write('строка 2\n')

# writelines - запись списка строк
f.writelines(['строка 3\n', 'строка 4\n'])

# close - закрытие файла
f.close()

# open - открытие для чтения
f = open('test.txt', 'r')

# read - чтение всего файла
print(f.read())

# readline - чтение одной строки
f.seek(0)
print(f.readline(), end='')

# readlines - чтение всех строк в список
f.seek(0)
print(f.readlines())

f.close()
````

````python
name = "Мария"
age = 20
print(f"Меня зовут {name}, мне {age} лет")
````



### Глава 9. Файлы

**Вопрос:** Какие режимы открытия файла существуют?

**Ответ:** 'r' (чтение), 'w' (запись), 'a' (добавление).

````python
# Запись
f = open('test.txt', 'w')
f.write('Привет')
f.close()

# Чтение
f = open('test.txt', 'r')
print(f.read()) 
f.close()

# Добавление
f = open('test.txt', 'a')
f.write(' Мир')
f.close()
````
