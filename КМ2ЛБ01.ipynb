## Вывод уравнений движения тела

### 1. Начальные условия

Поместим начало прямоугольной декартовой системы координат на поверхность земли. Ось $OX$ направим горизонтально, ось $OY$ — вертикально вверх.

В начальный момент времени $t = 0$ тело находится в точке с координатами:

$$
s_x(0) = 0, \quad s_y(0) = h_{\text{start}}
$$

где $h_{\text{start}}$ — начальная высота тела.

### 2. Второй закон Ньютона

Запишем второй закон Ньютона в векторной форме:

$$
m \vec{a} = \vec{F}
$$

где:
- $m$ — масса тела,
- $\vec{a}$ — ускорение тела,
- $\vec{F}$ — сила, действующая на тело.

### 3. Сила, действующая на тело

В данной задаче на тело действует только сила тяжести. В выбранной системе координат она направлена вертикально вниз и равна:

$$
\vec{F} = m(0, -g)
$$

где $g$ — ускорение свободного падения.

### 4. Ускорение тела

Ускорение тела есть вторая производная от перемещения по времени:

$$
\vec{a} = \left( s_x''(t), \, s_y''(t) \right)
$$

### 5. Подстановка во второй закон Ньютона

Подставляем выражения для силы и ускорения во второй закон Ньютона:

$$
m \left( s_x''(t), \, s_y''(t) \right) = m (0, -g)
$$

Сокращаем обе части на массу $m$ (так как $m \neq 0$):

$$
\left( s_x''(t), \, s_y''(t) \right) = (0, -g)
$$

### 6. Система дифференциальных уравнений

Из полученного векторного равенства следуют два независимых уравнения:

$$
\begin{cases}
s_x''(t) = 0 \\
s_y''(t) = -g
\end{cases}
$$

### 7. Решение уравнений

#### 7.1. Решение для горизонтальной координаты $s_x(t)$

Уравнение $s_x''(t) = 0$ означает, что ускорение по горизонтали равно нулю, следовательно, скорость постоянна.

Интегрируем первый раз:

$$
s_x'(t) = v_{0x} = \text{const}
$$

Интегрируем второй раз:

$$
s_x(t) = v_{0x} t + C
$$

где $C$ — постоянная интегрирования.

Используем начальное условие $s_x(0) = 0$:

$$
s_x(0) = v_{0x} \cdot 0 + C = 0 \quad \Rightarrow \quad C = 0
$$

Таким образом:

$$
\boxed{s_x(t) = v_{0x} t}
$$

#### 7.2. Решение для вертикальной координаты $s_y(t)$

Уравнение $s_y''(t) = -g$ означает, что ускорение по вертикали постоянно и направлено вниз.

Интегрируем первый раз:

$$
s_y'(t) = -g t + v_{0y}
$$

где $v_{0y}$ — вертикальная составляющая начальной скорости.

Интегрируем второй раз:

$$
s_y(t) = -\frac{g}{2} t^2 + v_{0y} t + D
$$

где $D$ — постоянная интегрирования.

Используем начальное условие $s_y(0) = h_{\text{start}}$:

$$
s_y(0) = -\frac{g}{2} \cdot 0^2 + v_{0y} \cdot 0 + D = h_{\text{start}} \quad \Rightarrow \quad D = h_{\text{start}}
$$

Таким образом:

$$
\boxed{s_y(t) = h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2}
$$

### 8. Итоговый аналитический вид функции перемещения

Объединяя полученные выражения, получаем функцию перемещения тела в векторном виде:

$$
\boxed{\vec{s}(t) = \bigl( v_{0x} t, \ h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2 \bigr)}
$$

где:
- $v_{0x}$ и $v_{0y}$ — проекции начальной скорости на оси координат,
- $h_{\text{start}}$ — начальная высота тела,
- $g$ — ускорение свободного падения,
- $t$ — время.

### Определение углового коэффициента касательной

Угловой коэффициент касательной к траектории движения тела в начальный момент времени $t = 0$ определяется как отношение вертикальной составляющей начальной скорости к горизонтальной составляющей начальной скорости:

$$k = \frac{v_{0y}}{v_{0x}}$$

где:
- $v_{0x} = \frac{s_{end}}{T}$ — горизонтальная составляющая начальной скорости,
- $v_{0y} = \frac{h_{end} - h_{start} + \frac{g}{2}T^2}{T}$ — вертикальная составляющая начальной скорости.

Для данной задачи:
- $v_{0x} = \frac{5}{3} = 1.6667$ м/с,
- $v_{0y} = \frac{3 - 1 + \frac{9.80665}{2} \cdot 3^2}{3} = 15.3766$ м/с.

Таким образом:
$$k = \frac{15.3766}{1.6667} = 9.2260$$
