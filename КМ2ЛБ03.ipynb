```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def x_param(t):
    return np.sqrt(1 - t**2)

def y_param(t):
    return t * np.tan(t + 1)

t_min = 0
t_max = 1
step_t = 0.01

t_array = np.arange(t_min, t_max, step_t)
x_array_t = x_param(t_array)
y_array_t = y_param(t_array)

matrix_t = np.transpose([x_array_t, y_array_t])

fig_t = plt.figure()
x_min_t, x_max_t = np.min(x_array_t), np.max(x_array_t)
y_min_t, y_max_t = np.min(y_array_t), np.max(y_array_t)
ax_t = plt.axes()
plt.axis([x_min_t-0.5, x_max_t+0.5, y_min_t-0.5, y_max_t+0.5])
line_t, = ax_t.plot([], [], 'b', lw=2)

def at_frame_t(point):
    x_coord = list(line_t.get_xdata())
    y_coord = list(line_t.get_ydata())
    x_coord.append(point[0])
    y_coord.append(point[1])
    line_t.set_data(x_coord, y_coord)
    return line_t,

anim = FuncAnimation(fig_t, at_frame_t, frames=matrix_t, repeat=False, interval=30)
plt.show()
```

## Задание 3.4


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def x(t):
    return np.sqrt(1 - t**2)

def y(t):
    return t * np.tan(t + 1)

t = np.arange(0, 1, 0.01)
x_arr = x(t)
y_arr = y(t)

idx = np.argsort(x_arr)
x_arr = x_arr[idx]
y_arr = y_arr[idx]

# Точка B
xB, yB = x_arr[-1], y_arr[-1]

# Численное дифференцирование для касательной
dt = 0.001
dx = (x(0.999) - x(0.999 - dt)) / dt
dy = (y(0.999) - y(0.999 - dt)) / dt

s = np.linspace(-2, 2, 100)
x_tan = xB + dx * s
y_tan = yB + dy * s

# График
fig, ax = plt.subplots()
ax.plot(x_arr, y_arr, 'b')
ax.plot(xB, yB, 'go')
ax.plot(x_tan, y_tan, 'r')
ax.set_xlim(min(x_arr)-0.5, max(x_arr)+0.5)
ax.set_ylim(min(y_arr)-0.5, max(y_arr)+0.5)

pointA, = ax.plot([], [], 'ro')
secant, = ax.plot([], [], 'g--')

def init():
    pointA.set_data([], [])
    secant.set_data([], [])
    return pointA, secant

def animate(frame):
    xA, yA = frame
    pointA.set_data([xA], [yA])
    secant.set_data([xA, xB], [yA, yB])
    return pointA, secant

matrix = np.transpose([x_arr, y_arr])
anim = FuncAnimation(fig, animate, frames=matrix, init_func=init, interval=30)
plt.show()
```
